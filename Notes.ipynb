## corpus

A corpus is simply a collection of text documents that you use in NLP. Think of it as the entire library of text your model can learn from (the training data).

### Example 1
Suppose you have these three sentences:

- I love machine learning.
- PyTorch is fun.
- I love Python.

The corpus is all three sentences together.
```
Corpus
├── I love machine learning.
├── PyTorch is fun.
└── I love Python.
```
